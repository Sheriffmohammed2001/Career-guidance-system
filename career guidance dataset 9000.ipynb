## analysing catogarical variables

## Label encoding

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 

df2['Database Fundamentals'] = label_encoder.fit_transform(df2['Database Fundamentals'])
df2['Computer Architecture'] = label_encoder.fit_transform(df2['Computer Architecture'])
df2['Distributed Computing Systems'] = label_encoder.fit_transform(df2['Distributed Computing Systems'])
df2['Cyber Security'] = label_encoder.fit_transform(df2['Cyber Security'])
df2['Networking'] = label_encoder.fit_transform(df2['Networking'])
df2['Software Development'] = label_encoder.fit_transform(df2['Software Development'])
df2['Programming Skills'] = label_encoder.fit_transform(df2['Programming Skills'])
df2['Project Management'] = label_encoder.fit_transform(df2['Project Management'])
df2['Computer Forensics Fundamentals'] = label_encoder.fit_transform(df2['Computer Forensics Fundamentals'])
df2['Technical Communication'] = label_encoder.fit_transform(df2['Technical Communication'])
df2['AI ML'] = label_encoder.fit_transform(df2['AI ML'])
df2['Software Engineering'] = label_encoder.fit_transform(df2['Software Engineering'])
df2['Business Analysis'] = label_encoder.fit_transform(df2['Business Analysis'])
df2['Communication skills'] = label_encoder.fit_transform(df2['Communication skills'])
df2['Data Science'] = label_encoder.fit_transform(df2['Data Science'])
df2['Troubleshooting skills'] = label_encoder.fit_transform(df2['Troubleshooting skills'])
df2['Graphics Designing'] = label_encoder.fit_transform(df2['Graphics Designing'])
df2['Role'] = label_encoder.fit_transform(df2['Role'])
df2.head()
```

## Train test split

```python
X = df2.drop(['Role'], axis=1)
y = df2['Role']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 999)
X_train.shape, X_test.shape
```

## Gausian naive base

```python

from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)

y_pred[:15]
len(y_pred)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
y_pred_train = gnb.predict(X_train)
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

## confusion matrix

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf = DecisionTreeClassifier(criterion="gini", max_depth=15, random_state=999)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

# Predict the response for test dataset
y_pred2 = clf.predict(X_test)
```

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred2)))
```

```python
y_pred_train2 = clf.predict(X_train)
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train2)))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
model = KNeighborsClassifier(n_neighbors=7)
model.fit(X_train, y_train) 
result = model.score(X_test, y_test)

print('Model accuracy score: {0:0.4f}'. format(result))
```

## SVM

```python
from sklearn.svm import SVC
from sklearn import metrics
model1 = SVC()
model1.fit(X_train, y_train) 
result = model1.score(X_test, y_test)

print('Model accuracy score: {0:0.4f}'. format(result))
```

```python
linear_svc=SVC(kernel='linear', C=1.0) 


# fit classifier to training set
linear_svc.fit(X_train,y_train)


# make predictions on test set
y_pred_test=linear_svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with linear kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
# Run SVM with polynomial kernel and C=1.0
# instantiate classifier with polynomial kernel and C=1.0
poly_svc=SVC(kernel='poly', C=1.0) 


# fit classifier to training set
poly_svc.fit(X_train,y_train)


# make predictions on test set
y_pred=poly_svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with polynomial kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
